```python
#Dependencies
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
#CSV files read into pandas dataframes
data = pd.read_csv("./Data.csv")
data.head()
```

```python
grouped_data1 = data.groupby(["Country","State"])["Date"].count()
grouped_data= grouped_data1.sort_values()
```

```python
ax = grouped_data.plot(kind='barh', title ="Sightings Per State",figsize=(15,10),legend=True, fontsize=12)
ax.set_xlabel("State",fontsize=12)
ax.set_ylabel("Number of Sightings",fontsize=12)
plt.savefig("./Sightings Per State", bbox_inches="tight")
```

```python
grouped_date1 = data.groupby(["Date"])["Date"].count()
grouped_date = grouped_date1.sort_values()
```

```python
ax = grouped_date.plot(kind='barh', title ="Sightings Per Day",figsize=(15,10),legend=True, fontsize=12)
ax.set_xlabel("State",fontsize=12)
ax.set_ylabel("Number of Sightings",fontsize=12)
plt.savefig("./Sightings Per Day", bbox_inches="tight")
```

```python
grouped_country = data.groupby(["Country"])["Date"].count()
grouped_country
```

```python
colors = ["#3ccedf","#ff3208"]
grouped_country.Date.value_counts().plot(kind='pie', autopct='%1.0f%%', pctdistance=0.5,shadow=True,legend=True,colors=colors)
plt.title("UFO sightings By Country")
plt.savefig("SightingsByCountry.png")
```
